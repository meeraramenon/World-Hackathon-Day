```r
#to display a density graph
Student %>%
  select(Chapter.4.Score) %>%                     # Keep only numeric columns
  gather() %>%                             # Convert to key-value pairs
  ggplot(aes(value)) +                     # Plot the values
  facet_wrap(~ key, scales = "free") +   # In separate panels
  geom_density()   
```

```r
# to display the pie chart of each student and their score

pie(Student$Chapter.4.Score,labels=paste(Student$Name.of.student,Student$Chapter.4.Score))
```
